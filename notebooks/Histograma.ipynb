```python
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imframe = Image.open('swimsegA.png')
```

```python
npframe = np.array(imframe.getdata())
imgrgbdf = pd.DataFrame(npframe)
imgrgbdf=imgrgbdf.rename(columns={0:"R",1:"G",2:"B"})
imgrgbdf #una fila per pixel, la imatge és de 110x100 pixels
```

```python
#histograma de R
plt.hist(imgrgbdf.R)
```

```python
#histograma de G
plt.hist(imgrgbdf.G)
```

```python
#histograma de B
plt.hist(imgrgbdf.B)
```

```python
#reconvertim 
imgrgbdf['Rnou'] = np.where(imgrgbdf['R']>150, 0, 255)
imgrgbdf['Gnou'] =imgrgbdf['Rnou'] 
imgrgbdf['Bnou'] =imgrgbdf['Rnou'] 
npModificat=imgrgbdf[['Rnou','Gnou','Bnou']].to_numpy()
npModificat=npModificat.reshape(110,110,3)
im3=Image.fromarray(npModificat.astype(np.uint8))
plt.imshow(im3)
```
